```python
import re #regular expression
```

```python
text = [
    'the rain is blue, %thinking about the future is good.% what color is the rain? blue.',
    'the cat is yellow, %deep into the computer science.% what color is the cat? yellow.',
    'the flower is red, %hate to stay.% what color is the flower? red.',
    'the sky is purple, %the data is position.% what color is the sky? purple.',

    'the apple is green, %sometimes people forget what really matters in life.% what color is the apple? green.',
    'the car is black, %running through endless roads without direction feels strange.% what color is the car? black.',
    'the dog is white, %coding all night makes the brain feel heavy but satisfied.% what color is the dog? white.',
    'the ocean is blue, %there are many theories about intelligence but few about wisdom.% what color is the ocean? blue.',

    'the chair is brown, %short thought.% what color is the chair? brown.',
    'the pen is red, %this is a much longer filler sentence that contains multiple clauses and tries to distract the model from the original fact by adding complexity and length.% what color is the pen? red.',
    'the phone is silver, %randomness increases when the sequence becomes longer and harder to track.% what color is the phone? silver.',

    'the bird is orange, %walking quietly.% what is the color of the bird? orange.',
    'the book is blue, %sometimes sentences are irrelevant and meaningless.% what is the color of the book? blue.',
    'the table is gray, %even if the filler talks about science mathematics or philosophy it should not matter.% what is the table color? gray.',

    'the banana is yellow, %AI models tend to overfit if patterns are too simple.% tell me the color of the banana. yellow.',
    'the shirt is pink, %adding noise helps generalization in many cases.% tell me the color of the shirt. pink.',

    'the mountain is white, %snow covers everything and silence dominates the landscape.% what color is the mountain? white.',
    'the river is green, %long long filler text that keeps going without giving useful information just to test robustness and attention span.% what color is the river? green.',

    'the cup is black, %irrelevant words irrelevant words irrelevant words irrelevant words.% what color is the cup? black.',
    'the lamp is yellow, %mixing short and long fillers improves dataset diversity.% what color is the lamp? yellow.',

    'the leaf is green, %first clause meaningless second clause also meaningless third clause still meaningless.% what color is the leaf? green.',
    'the stone is gray, %distractions can appear before or after the key information and models must learn to focus.% what color is the stone? gray.'
]
```

```python
class TrainingData:
    def __init__(self, text) -> None:
        self.text = text
        self.token = self.tokenization()
        self.dist = self.distance()

    def tokenization(self):
        result_ = []
        for s in self.text:
            tokens = re.findall(r"%|[a-zA-Z]+|[^\w\s]", s)
            result_.append(tokens)
        return result_

    def distance(self):
        dist_ = []
        for sample in self.token:
            if sample.count('%') < 2:
                dist_.append(0)
                continue
            first = sample.index('%')
            second = sample.index('%', first + 1)
            dist_.append(second-first)
        return dist_
    
TraingData = TrainingData
# train_data = TraingData(text[0:2])
train_data = TraingData(text)
```

```python
train_data.dist[0:5]
```

```python
vocabulary = {}
idx = 0
for sample in train_data.token:
    for words in sample:
        if words not in vocabulary:
            vocabulary[words] = idx
        idx += 1
    
```
