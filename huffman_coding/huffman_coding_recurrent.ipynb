编写一个 Python 脚本，要求：

输入：一个字典，键是字符，值是概率（加和为 1）。

输出：每个字符对应的哈夫曼编码（0 和 1 字符串）。

🛠️ 规定工具箱（限制条件）为了锻炼你的逻辑，我规定你必须使用以下工具：
heapq 模块：这是 Python 自带的“最小堆”工具。为什么要用它？ 因为哈夫曼算法每次都要找“最小的两个数”。如果用普通的 list.sort()，每次合并都要重新排序，效率太低。heapq 可以让你以 $O(\log n)$ 的效率自动维护最小元素。

class (类)：定义一个 Node 类。这个类至少要有四个属性：char (字符), freq (概率), left (左孩子), right (右孩子)。递归函数：用于最后遍历整棵树生成编码。

```python
data = {
    'TEMP': 0.40,
    'HUMI': 0.25,
    'PRES': 0.20,
    'WIND': 0.10,
    'RAIN': 0.05
}
```

```python
import heapq

class Node:
    def __init__(self, char, freq): 
        self.char = char
        self.freq = freq
        self.left = None
        self.right = None
    def __lt__(self, other): #heapq的两两比较protocol函数
        return self.freq < other.freq

def build_huffman_tree(data): # return root node
    nodes = [Node(char, freq) for char, freq in data.items()]
    heapq.heapify(nodes) #最小堆两两比较

    while len(nodes) > 1:
        left_node = heapq.heappop(nodes)
        right_node = heapq.heappop(nodes)

        new_node = Node(None, left_node.freq+right_node.freq)
        new_node.left = left_node
        new_node.right = right_node

        heapq.heappush(nodes, new_node)
    return nodes[0]

def generate_codes(node, current_code, codes): #我讨厌递归
    if node is None:
        return codes
    if node.char is not None:
        codes[node.char] = current_code
        return
    
    generate_codes(node.left, current_code + '0', codes)
    generate_codes(node.right, current_code + '1', codes)
```

```python
if __name__ == "__main__":
    root = build_huffman_tree(data)
    huffman_code = {}
    generate_codes(root, '', huffman_code)
    print('The huffman codes:')
    for char, code in huffman_code.items():
        print(f"{char}: {code}")
```
