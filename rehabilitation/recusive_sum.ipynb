```python
def recursive_sum(lst):
    if lst==[]:
        return 0
    return recursive_sum(lst[0:-1]) + lst[-1]
recursive_sum([1, 2, 3, 4])
```

```python
def recursive_reverse(lst):
    if len(lst) == 1:
        return lst
    return recursive_reverse(lst[1:]) + [lst[0]]

recursive_reverse([1, 2, 3, 4])
```

```python
def flatten(lst):
    if lst == []:
        return lst
    if type(lst[0]) is list:
        return flatten(lst[0])
    else: 
        return flatten(lst[1:])
     
    
# flatten([1, [2, 3], [4, [5, 6]]])
flatten([[1, 2], [3, 4]])
```
