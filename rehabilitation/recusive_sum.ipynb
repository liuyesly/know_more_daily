```python
def recursive_sum(lst):
    if lst==[]:
        return 0
    return recursive_sum(lst[0:-1]) + lst[-1]
recursive_sum([1, 2, 3, 4])
```

```python
def recursive_reverse(lst):
    if len(lst) == 1:
        return lst
    return recursive_reverse(lst[1:]) + [lst[0]]

recursive_reverse([1, 2, 3, 4])
```

```python
def power(a, b):
    if b == 0:
        return 1
    return power(a, b-1) * a

power(3, 4)
        
```

```python
def count(lst, target):
    if lst == []:
        return 0
    
    if lst[0] == target:
        return count(lst[1:], target) + 1
    else:
        return count(lst[1:], target) + 0
    

count([1, 2, 3, 2, 2], 2)
# count([1, 2, 3], 5)
```

```python
class TreeNode():
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None
```

```python
root = TreeNode(1)
root.left = TreeNode(2)
root.right = TreeNode(3)
root.left.left = TreeNode(4)
root.left.right = TreeNode(5)
```

```python
def tree_sum(root):
    if root == None:
        return 0
    return root.val + tree_sum(root.left) + tree_sum(root.right)
tree_sum(root)
```

```python
def tree_depth(root):
    if root == None:
        return 0
    return 1 + max(tree_depth(root.left), tree_depth(root.right))
tree_depth(root)
```

```python
def countLeaves(root):
    if root == None:
        return 0
    if root.left == None and root.right == None:
        return 1
    return countLeaves(root.left) + countLeaves(root.right)
countLeaves(root)
```

```python
def has_path_sum(root, target):
    if root is None:
        return False
    if root.left is None and root.right is None:  # 叶子节点
        return target == root.val
    return has_path_sum(root.left, target - root.val) or \
           has_path_sum(root.right, target - root.val)
has_path_sum(root, 6)
```

```python
def display(root, level=0, prefix="Root: "):
    if root is None:
        return
    print(" " * (level * 4) + prefix + str(root.val))
    if root.left or root.right:
        if root.left:
            display(root.left, level + 1, "L: ")
        else:
            print(" " * ((level + 1) * 4) + "L: None")
        if root.right:
            display(root.right, level + 1, "R: ")
        else:
            print(" " * ((level + 1) * 4) + "R: None")
display(root)
```

```python
def invert(root):
    if root is None:
        return None
    root.left, root.right = root.right, root.left
    invert(root.left)
    invert(root.right)
invert(root) 
display(root)  
```

```python
def lca(root, target1, target2):
    if root.left and root.right == None: # if leaf
        if root.val in (target1, target2):
            return root.val
        else:
            return None
    
lca(root, 4, 3)
    
```
