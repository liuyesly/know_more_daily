```python
class Value():
    def __init__(self, val):
        self.val = val
        self.grad = 0.0
        self._children = None
        self._backward = lambda: None # 提前声明，很重要！
    def __repr__(self):
        return f'Value({self.val})'
    def __mul__(self, other):
        ans = Value(self.val * other.val)
        ans._children = (self, other)
        def _backward():
            self.grad = other.val * ans.grad
            other.grad = self.val * ans.grad
        ans._backward = _backward # stores the function itself as a method you may make use of. 
        return ans
    def __add__(self, other):
        ans = Value(self.val + other.val)
        ans._children = (self, other)
        def _backward():
            self.grad = 1.0 * ans.grad
            other.grad = 1.0 * ans.grad
        ans._backward = _backward
        return ans
    
    def backward(self):
        print(f'the backward root {self.val}, and grad: {self.grad}')
        self._backward()
        if self._children is not None:
            if self._children[0] is not None: # if root node
                self._children[0].backward()
            if self._children[1] is not None: # if root node
                self._children[1].backward()
        pass
        
```

```python
# c = a * b
a = Value(2.0)
b = Value(3.0)
c = a * b
d = c * Value(4.0)
d.grad = 1.0
d.backward()
print(a.grad)
print(b.grad)
```
