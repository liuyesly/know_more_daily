```python
data = {
    'TEMP': 0.40,
    'HUMI': 0.25,
    'PRES': 0.20,
    'WIND': 0.10,
    'RAIN': 0.05
}
```

```python
import heapq

class Node:
    def __init__(self, char, freq): 
        self.char = char
        self.freq = freq
        self.left = None
        self.right = None
    def __lt__(self, other): #heapq的两两比较protocol函数
        return self.freq < other.freq

def build_huffman_tree(data): # return root node
    nodes = [Node(char, freq) for char, freq in data.items()]
    heapq.heapify(nodes) #最小堆两两比较

    while len(nodes) > 1:
        left_node = heapq.heappop(nodes)
        right_node = heapq.heappop(nodes)

        new_node = Node(None, left_node.freq+right_node.freq)
        new_node.left = left_node
        new_node.right = right_node

        heapq.heappush(nodes, new_node)
    return nodes[0]

def generate_codes_iterative(root, huffman_code): #非递归写法
    stack = [(root, "")]

    while stack:
        node, current_code = stack.pop()
        if node.char != None: # 叶子节点，可以赋码了
            huffman_code[node.char] = current_code
        else:
            stack.append((node.right, current_code + "1"))
            stack.append((node.left, current_code + "0"))
            
```

```python
if __name__ == "__main__":
    root = build_huffman_tree(data)
    huffman_code = {}
    generate_codes_iterative(root, huffman_code)
    print('The huffman codes:')
    for char, code in huffman_code.items():
        print(f"{char}: {code}")
```

```python
def decode(root, encoded_str):
    decode_output = []
    curr = root
    for bit in encoded_str:
        if bit =='0':
            curr = curr.left
        else:
            curr = curr.right
        
        if curr.char != None:
            decode_output.append(curr.char)
            curr = root
            
    return "".join(decode_output)

encoded_str = '0000'
print('The decode is:')
print(decode(root, encoded_str))
```
