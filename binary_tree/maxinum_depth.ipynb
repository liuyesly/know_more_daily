```python
data = [3, 9, 20, 15, 7, 30, 50, 90, 100]
```

```python
class TreeNode:
    def __init__(self, val=0, left=None, right = None):
        self.val = val
        self.left = left
        self.right = right
    def be_father(self, other, pos): #pos = 0, left; 1, right
        if pos == 0:
            self.left = other
        elif pos == 1:
            self.right = other
        else:
            print('pos = 0 for right child, 1 for left child.')
```

```python
# list实现，不好找node，换成字典
all_nodes = []
idx = 0
for value in data:
    node = TreeNode(val = value, left=None, right=None)
    all_nodes.append(node)
    idx +=1
# hard code the binary tree
all_nodes[0].be_father(all_nodes[1], 0)
all_nodes[0].be_father(all_nodes[2], 1)
all_nodes[2].be_father(all_nodes[3], 0)
all_nodes[2].be_father(all_nodes[4], 1)
```

```python
# 有格式的print
def print_binary_tree(root):
    # print("The root is now" + str(root.val))
    if root is None:
        return ""
    
    left_str = print_binary_tree(root.left)
    right_str = print_binary_tree(root.right)
    current = str(root.val)

    if left_str or right_str:
        result = current + "\n"
        
        if left_str:
            result += "|-- " + left_str.replace("\n", "\n|   ") + "\n"
        if right_str:
            result += "|-- " + right_str.replace("\n", "\n|   ")
    else:
        result = current
    
    return result
root = all_nodes[3]
tree_str = print_binary_tree(root)
print(tree_str)
```

```python
class Solution:
    def maxDepth(self, root: TreeNode):
        if root is None:
            return -1 #如果是-1就是算edge数量, 0按node来
        count_left = self.maxDepth(root.left)
        count_right = self.maxDepth(root.right)

        count = 1 + max(count_left, count_right)
        return count     
    
    def isBalanced_full_search(self, root: TreeNode): # 这种写法无法提前跳出，只能遍历完整颗树
        if root is None:
            return -1 
        count_left = self.isBalanced_full_search(root.left)
        count_right = self.isBalanced_full_search(root.right)
        count = 1 + max(count_left, count_right)

        diff = abs(count_left - count_right)
        # print(f'at root {root.val}, left,{count_left} right, {count_right} diff: {diff}')

        if diff>1:
            print('The tree is not balanced. ')
            return False   
        return count

    # def isBalanced(self, root: TreeNode):
    #     def dfs(root):
    #         if root.val is None:
    #             return 0
    #         height = root.max_depth(root)
    #         print(height)

    #     return 1
    #     # return dfs(root) != -1
    def isBalanced(self, root: TreeNode):
        if root is None:
            return -1
        height_left = self.maxDepth(root.left)
        height_right = self.maxDepth(root.right)
        if abs(height_right-height_left)>1:
            return False
        # print(height_left, height_right)

sol = Solution()
# print(sol.maxDepth(root))
print(sol.isBalanced(root))
```
